```python
#GV 25.Sep-2025
import sys
sys.path.append('amos') 
from amos.midi2df2midi import midi_to_dataframe, save_midi_from_df
from amos.midi2cube2midi import midi_to_cubematrix, cubematrix2df, save_midi, plot_pianoroll_part
import numpy as np
```

```python
#Testing df2cubematrix and cubematrix2df GV
filein='midis/sugar-plum-fairy_orch.mid' #Original file
fileout='midis/reconstructed2.mid' #Recunstructed file from cubematrix
#Convert midi to cube matrix
pianoroll_o, tcp_to_indexo, index_to_tcp, tcp_to_tracknameo = midi_to_cubematrix(filein, samples_per_qn=12)
```

```python
# Invert the cube back to midi
dfnmat = cubematrix2df(pianoroll_o, samples_per_qn = 12, tcp_to_index = tcp_to_indexo, index_to_tcp=index_to_tcp, tcp_to_trackname=tcp_to_tracknameo) 
```

```python
#save dfnmat to fileout
save_midi(dfnmat,fileout)
```
